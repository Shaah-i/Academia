# Largest palindrome product

```python
r=list(str(999*982))
r
```

```python
l=[]
for i in range(10,100):
    for j in range(10,100):
        r = list(str(i*j))
        if r[0]==r[-1] and r[1]==r[-2] and r[2]==r[-3]:
            m=int("".join(r))
            l.append([i,j,m])
print(l)
```

```python
count = 0
for i in range(100,10,-1):
    for j in range(100,10,-1):
        r = list(str(i*j))
        if r[0]==r[-1] and r[1]==r[-2] and r[2]==r[-3]:
            m="".join(r)
            print(f"{i}*{j} = {m}")
            count += 1
        if count > 0:
            break
```

```python
count = 0
for i in range(1000,100,-1):
    for j in range(1000,100,-1):
        r = list(str(i*j))
        if r[0]==r[-1] and r[1]==r[-2] and r[2]==r[-3] and len(r)==6:
            m="".join(r)
            print(f"{i}*{j} = {m}")
            count += 1
        if count > 0:
            break
```

```python
l=[]
for i in range(100,1000):
    for j in range(100,1000):
        r = list(str(i*j))
        if r[0]==r[-1] and r[1]==r[-2] and r[2]==r[-3] and len(r)==6:
            m="".join(r)
            l.append([m])
print(max(l))
```

```python
from scipy.stats import shapiro,stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
URL = ('https://drive.google.com/file/d/1mCjtYHiX--mMUjicuaP2gH3k-SnFxt8Y/view?usp=share_')
file_id=URL.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
dwn_url
```

```python
df= pd.read_csv(dwn_url, index_col="Patient ID")
df.head()
```

```python
df.isna().sum()
```

```python
stat,pvalue=shapiro(df)
```

```python
print('stat',stat)
print('pvalue', pvalue)
```

```python
for col in df.columns:
    sns.distplot(df[col],hist=True,kde=True,label=col)
    plt.title('Distribution Plot for Each Column')
    plt.xlabel(col)
    plt.ylabel('Density')

    plt.show()
```

### 4. Pokemons that do not have more than 2 evolutions

```python
df1 = df.copy()
```

```python
df1['next_evolution'].dropna(inplace= True)
```

```python
import ast
```

```python
df1["Evloutions_2"] = df1["next_evolution"].dropna().apply(lambda x : ast.literal_eval(x)).apply(lambda x : len(x)) < 2
```

```python
df1.loc[df1["Evloutions_2"] == True]["name"]
```

### 5. Pokemons whose spawn time is less than 300 seconds

```python
df1["spawn_time_less_then_300"] = df1["spawn_time"].dropna().apply(lambda x: (int(x.split(":")[0]) * 60) +(int(x.split(":")[1]))) < 300
```

```python
df1.loc[df1["spawn_time_less_then_300"] == True]["name"]
```

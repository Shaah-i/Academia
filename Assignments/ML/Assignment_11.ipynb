### 1.	Given X be a discrete random variable with the following PMF


 
    1. Find the range RX of the random variable X.

    2. Find P(X ≤ 0.5)

    3. Find P(0.25<X<0.75)

    4. P(X = 0.2|X<0.6)

**Ans.**

1. The range of X can be found from the PMF. The range of X consists of possible values for X. Here we have</br>
$RX={0.2,0.4,0.5,0.8,1}$</br>


2. The event X≤0.5 can happen only if X is 0.2,0.4, or 0.5. Thus,</br>
$P(X≤0.5)	=P(X∈{0.2,0.4,0.5})$</br>
$=P(X=0.2)+P(X=0.4)+P(X=0.5)$</br>
$=PX(0.2)+PX(0.4)+PX(0.5)$</br>
$=0.1+0.2+0.2=0.5$</br>


3. The event 0.25<X<0.75 can happen only if X is 0.4 or 0.5. Thus,</br>
$P(0.25<X<0.75)	=P(X∈{0.4,0.5})$</br>
$=P(X=0.4)+P(X=0.5)$</br>
$=PX(0.4)+PX(0.5)$</br>
$=0.2+0.2=0.4$</br>


4. This is a conditional probability problem, so we can use our famous formula P(A|B)=P(A∩B)P(B). We have</br>
$P(X=0.2|X<0.6)	=P((X=0.2) and (X<0.6))P(X<0.6)$</br>
$=P(X=0.2)P(X<0.6)$</br>
$=PX(0.2)PX(0.2)+PX(0.4)+PX(0.5)$</br>
$=0.10.1+0.2+0.2=0.2$</br>

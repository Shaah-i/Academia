### 1. Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave


```python
performance['Date']  = pd.to_datetime(performance['Date'])
performance.info()
```

```python
# Series.dt.weekofyear and Series.dt.week have been deprecated.
performance['Week'] = performance['Date'].dt.isocalendar().week
performance
```

```python
weekly_average_rating = pd.DataFrame(performance.groupby(['Agent Name','Week'])['Average Rating'].mean())
weekly_average_rating
```

### 2 .Total working days for each agents

```python
performance['Day'] = performance['Date'].dt.day
performance
```

```python
working_days = pd.DataFrame(performance.groupby(['Agent Name'])['Day'].count())
working_days
```

### 3. Total query that you have taken.

```python
Total_queries = pd.DataFrame(performance.groupby('Agent Name')['Total Chats'].sum())
Total_queries
```

### 4. Total Feedback that you have received

```python
Total_feedbacks = pd.DataFrame(performance.groupby('Agent Name')['Total Feedback'].sum())
Total_feedbacks
```

### 5. An agent name who have average rating between 3.5 to 4

### 9. Average weekly response time for each agent

### 10. Average weekely resolution time for each agents
